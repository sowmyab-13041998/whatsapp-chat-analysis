### Seperating into data columns

```python
pattern = '\d{1,2}\/\d{2,4}\/\d{2,4},\s\d{1,2}:\d{1,2}\s\w{1,2}\s-\s'
```

```python
messages =re.split(pattern, data)[1:]
len(messages)
```

```python
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
chat= pd.DataFrame({'user_message':messages,'message_date':dates})
chat['message_date'] =pd.to_datetime(chat['message_date'], format='%d/%m/%y, %I:%M %p - ')

chat.rename(columns={'message_date':'date'}, inplace = True)

chat.head()

                                     
```

```python
chat.shape
```

### Seperating user and the messages

```python
users=[]
mesages =[]
for message in chat['user_message']:
    entry=re.split('([\w\W]+?):\s',message)
    if entry[1:]: # user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('chat_notification')
        messages.append(entry[0])
chat['user'] = users
chat['message']=message
chat.drop(columns=['user_messages'],inplace = True)


        
```
